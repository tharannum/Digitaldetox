```python
l = list()
def fetch_question_text_and_weight(df):
    for i in range(len(df)):
        l.append((df.question_text.iloc[i], df['Weight (%)'].iloc[i]))
    print(l)
```

```python
fetch_question_text_and_weight(df)
```

```python
print('INSERT INTO questions (question_text, weightage) ')
print('VALUES ')
text = ''
num = 0
for i in l:
    text = i[0]
    num = int(i[1])
    print((text, num), end=',\n')
```

```python
a, b, c, d = '', '', '', ''
def fetch_options(df):
    for i in range(len(df)):
        a, b, c, d = df.option_a.iloc[i], df.option_b.iloc[i], df.option_c.iloc[i], df.option_d.iloc[i]
        print((int(df.Id.iloc[i]), 'A', a.strip()), end=',\n') 
        print((int(df.Id.iloc[i]), 'B', b.strip()), end=',\n') 
        print((int(df.Id.iloc[i]), 'C', c.strip()), end=',\n') 
        print((int(df.Id.iloc[i]), 'D', d.strip()), end=',\n') 
```

```python
print('INSERT INTO question_options (question_id, option_label, option_text)\nVALUES')
fetch_options(df)
```
